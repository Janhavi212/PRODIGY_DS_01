```python
import pandas as pd

# Read the CSV file, skipping metadata rows
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_85220.csv", skiprows=4)

# Show the first few rows
df.head()
```

```python
# Select columns (e.g., country name and population in 2020)
df_selected = df[["Country Name", "2020"]]

# Remove NaNs and sort
df_selected = df_selected.dropna().sort_values("2020", ascending=False)

# Optional: Take top 10 countries
top10 = df_selected.head(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(top10["Country Name"], top10["2020"], color="skyblue")
plt.xticks(rotation=45)
plt.title("Top 10 Most Populous Countries (2020)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df_selected["2020"], bins=30, color="orange", edgecolor="black")
plt.title("Histogram of Country Populations (2020)")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.show()
```
